#### EDA
***Task 5. Data Analysis***

**Univariate Analysis :**
***Analyzing each of the single variable likes single columns and their data distribution***

```python
# Identifying Outliers in price

df1 = df[df['price'] < 1500]

sns.boxplot(data=df1, x='price')
```

```python
# Price Distribution
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df1, x='price',bins=100,kde=True)
plt.title('Price Distribution')
plt.ylabel('frequency')
plt.show()
```

```python
df.columns
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df1, x='availability_365',kde=True)
plt.title('Availability Distribution')
plt.ylabel('frequency')
plt.show()
```

```python
df1.dtypes
```

```python
# Average price per bed
df1.groupby('neighbourhood_group')['price_per_bed'].mean()
```

```python
df1.groupby('neighbourhood_group')['price'].mean()
```

***Feature Engineering***

```python
#['price_per_bed']

df1['price_per_bed'] = df1['price']/ df1['beds']
df1.head()
```

***Bivariable Analysis : you will be analyze one column with another column***

```python
df1.columns
```

```python
# price dependency on neighbourhood_group
sns.barplot(data=df1, x='neighbourhood_group', y='price', hue='room_type')
```

```python
# number_of_reviews and price relationship
plt.figure(figsize=(8, 5))
plt.title('Locality and reviews dependency')
sns.scatterplot(data=df1, x='number_of_reviews', y='price', hue='neighbourhood_group')
plt.show()
```

```python
df1.dtypes
```

```python
sns.pairplot(data=df1, vars=['price','minimum_nights','number_of_reviews','availability_365'], hue='room_type')
```

```python
df1.columns
```

```python
# Geographical Distribution of AirBnB Listing
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df1, x='longitude', y='latitude', hue='room_type')
plt.title('Geographical Distribution of AirBnB Listing')
plt.show()
```

```python
df1.dtypes
```

```python
# Heat Map - Correlation of One Variable With Others for Numeric Column
corr = df1[['latitude','longitude','price','minimum_nights','number_of_reviews','reviews_per_month','availability_365','beds']].corr()
corr

plt.figure(figsize=(10, 6))
sns.heatmap(data=corr, annot=True)
```
